```python
from PIL import Image
import matplotlib.pyplot as plt
import time
import cv2
from google.colab.patches import cv2_imshow
```

#Original Image

```python
img = cv2.imread("tulip.jpg")
cv2_imshow(img)
```

#Thresholding in OpenCV

```python
# OpenCV
# By default BGR image
start_time = time.time()
img = cv2.imread("tulip.jpg")

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(f"Size of the original image : {img.shape[1]} x {img.shape[0]}")
print(f"Aspect ratio of original image: {img.shape[1]/img.shape[0]}")

ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)

plt.imshow(thresh1)
plt.show()

cv2_time = time.time()-start_time
print(f"Total time taken by cv2 is {cv2_time} seconds")
```

#Thresholding in PIl

```python
threshold = 127
image = Image.open("tulip.jpg")
#BINARY
image = image.point( lambda p: 255 if p > threshold else 0 )

PIL_time = time.time()-start_time
print(PIL_time)
plt.imshow(image)
plt.show()

print(f"Total time taken by pil is {PIL_time} seconds")
```
